```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Load train/test data 
X_train = pd.read_csv('X_train.csv')
X_test  = pd.read_csv('X_test.csv')
y_train = pd.read_csv('y_train.csv').squeeze()   # convert to 1D array
y_test  = pd.read_csv('y_test.csv').squeeze()


# train the model

model = LogisticRegression(max_iter=1000, random_state=42)

model.fit(X_train, y_train)    # train on training data
print("Model trained!")


# maing predictions on test data
y_pred = model.predict(X_test)


```

### MOdel evaluation

```python

# Accuracy 
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy: {accuracy:.4f}  ({accuracy*100:.1f}%)")

# Classification Report 
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['Toss Lost', 'Toss Won']))

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted Lost', 'Predicted Won'],
            yticklabels=['Actual Lost', 'Actual Won'])
plt.title('Logistic Regression - Confusion Matrix')
plt.tight_layout()
plt.savefig('ipl_lr_confusion_matrix.png', dpi=150)
plt.show()
print("Confusion matrix saved as 'ipl_lr_confusion_matrix.png'")
```
