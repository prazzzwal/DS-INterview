```python

import pandas as pd
df_clean = pd.read_csv('ipl_clean.csv')

df_feat = df_clean.copy()
```

```python
# feature 1: Did the toss winner win the match?
# if toss_winner == winner then value is 1 else 0

df_feat['toss_match_win'] = (df_feat['toss_winner'] == df_feat['winner']).astype(int)

print("Feature 1: 'toss_match_win' created")
print(df_feat[['toss_winner', 'winner', 'toss_match_win']].head())

```

```python

# feature 2: Win margin category
# Was the win Large, Medium, or Small?
# -----------------------------------------------
def get_win_category(row):
    if row['win_by_runs'] > 0:
        # Won by runs (batting team won)
        if row['win_by_runs'] >= 50:
            return 'Large'
        elif row['win_by_runs'] >= 20:
            return 'Medium'
        else:
            return 'Small'
    elif row['win_by_wickets'] > 0:
        # Won by wickets (chasing team won)
        if row['win_by_wickets'] >= 7:
            return 'Large'
        elif row['win_by_wickets'] >= 4:
            return 'Medium'
        else:
            return 'Small'
    else:
        return 'Tie or No Result'

df_feat['win_margin_cat'] = df_feat.apply(get_win_category, axis=1)

print("\nFeature 2: 'win_margin_cat' created")
print(df_feat['win_margin_cat'].value_counts())

```

```python

# feature 3: extracting year and month from match date
df_feat['date'] = pd.to_datetime(df_feat['date'], errors='coerce')
df_feat['match_year']  = df_feat['date'].dt.year
df_feat['match_month'] = df_feat['date'].dt.month

print("\nFeature 3: 'match_year' and 'match_month' extracted from date")
print(df_feat[['date', 'match_year', 'match_month']].head())
```

```python


# waas it a dominant win by runs? (>50 runs)
# flag: 1 = big win, 0 = not

df_feat['is_big_win'] = (df_feat['win_by_runs'] > 50).astype(int)

print("\nFeature 4: 'is_big_win' created")
print(df_feat['is_big_win'].value_counts())
```

```python
# Save for next sections
df_feat.to_csv('ipl_features.csv', index=False)
print("\nFeature-engineered data saved as 'ipl_features.csv'")
```
