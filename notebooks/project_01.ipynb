### Handling Missing Values

```python
df_clean = df.copy()

# Strategy 1: Drop rows where 'winner' is missing (only 4 rows - likely tied/no result)
print(f"\nRows before dropping missing 'winner': {len(df_clean)}")
df_clean = df_clean.dropna(subset=['winner'])
print(f"Rows after dropping missing 'winner': {len(df_clean)}")

```

```python
# strategy 2: filling 'city' with mode (most frequent city)
most_frequent_city = df_clean['city'].mode()[0]
df_clean['city'].fillna(most_frequent_city, inplace=True)
print(f"\n'city' missing values filled with mode: '{most_frequent_city}'")
```

```python

#strategy 3: filling 'player_of_match' with placeholder
df_clean['player_of_match'].fillna('Unknown', inplace=True)
print("'player_of_match' missing values filled with 'Unknown'")
```

```python

# strategy 4: filling umpire columns with unknown
df_clean['umpire1'].fillna('Unknown', inplace=True)
df_clean['umpire2'].fillna('Unknown', inplace=True)

```

```python
# strategy 5: Dropping 'umpire3' because it is  missing 637 rows in the dataset 
df_clean.drop(columns=['umpire3'], inplace=True)
print("umpire3 column dropped")

print("\nMissing values after treatment ")
print(df_clean.isnull().sum())
```

### Removing Duplicate rows

```python
print(f"\nTotal duplicate rows: {df_clean.duplicated().sum()}")
df_clean = df_clean.drop_duplicates()

# No any duplicate rows so doesnt need to remove anything
print(f"Dataset shape after removing duplicates: {df_clean.shape}")
```

### LAbel Encoding

```python

from sklearn.preprocessing import LabelEncoder
import pandas as pd

df_encoded = df_clean.copy()

# label encoder object
le = LabelEncoder()

df_encoded['toss_decision_enc'] = le.fit_transform(df_encoded['toss_decision'])

print("toss_decision encoding:")
print("bat   →", 0)
print("field →", 1)
print(df_encoded[['toss_decision', 'toss_decision_enc']].head())

```

### One Hot Encoding


```python
# One-Hot Encoding for categorical columns
ohe_cols = ['Season', 'city']  # columns with nominal data
df_encoded = pd.get_dummies(df_encoded, columns=ohe_cols, drop_first=True)

print(f"Shape after OHE: {df_encoded.shape}")
print("Sample new OHE columns:")
ohe_sample = [c for c in df_encoded.columns if c.startswith('city_') or c.startswith('season_')]
print(ohe_sample[:10])
```
