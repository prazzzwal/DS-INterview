## Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt

df_clean = pd.read_csv('ipl_clean.csv')
```

```python
# top 10 teams with most wins (Bar Chart) 
top_winners = df_clean['winner'].value_counts().head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_winners.index, top_winners.values, color='steelblue')
plt.title('Top 10 Teams by Wins', fontsize=14, fontweight='bold')
plt.xlabel('Number of Wins', fontsize=12)
plt.gca().invert_yaxis()   

# Add value labels on the bars
for i, v in enumerate(top_winners.values):
    plt.text(v + 0.5, i, str(v), va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
#number of Matches per Season (Line Chart)

matches_per_season = df_clean['Season'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.plot(matches_per_season.index,
         matches_per_season.values,
         marker='o', 
         color='purple',
         linewidth=2,
         markersize=8)
plt.title('Number of Matches per Season', fontsize=14, fontweight='bold')
plt.xlabel('Season', fontsize=12)
plt.ylabel('Number of Matches', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# Add value labels on each point
for i, (season, count) in enumerate(matches_per_season.items()):
    plt.text(season, count + 0.5, str(count), 
             ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# Scatter Plot: Runs vs Wickets Victory Comparison

plt.figure(figsize=(10, 6))

# Filter data
runs_wins_data = df_clean[df_clean['win_by_runs'] > 0]
wickets_wins_data = df_clean[df_clean['win_by_wickets'] > 0]

# Create scatter plots
plt.scatter(runs_wins_data['win_by_runs'], 
            [0] * len(runs_wins_data),  # Place on y=0
            alpha=0.6, color='coral', label='Won by Runs', s=60)

plt.scatter(wickets_wins_data['win_by_wickets'], 
            [1] * len(wickets_wins_data),  # Place on y=1
            alpha=0.6, color='mediumseagreen', label='Won by Wickets', s=60)

plt.title('Match Victory Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Win Margin', fontsize=12)
plt.yticks([0, 1], ['Win by Runs', 'Win by Wickets'])
plt.grid(axis='x', alpha=0.3)
plt.legend()
plt.show()
```
